# Missing values

```python
# First lets handle Categorical features which are missing
categorical_features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']
for feature in categorical_features_nan:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#Replace missing values with a new label
def replace_cat_feature(dataset,categorical_features_nan):
    data=dataset.copy()
    data[categorical_features_nan]= data[categorical_features_nan].fillna('Missing')
    return data
dataset=replace_cat_feature(dataset,categorical_features_nan)
dataset[categorical_features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
## Now lets check for numerical variables contain missing values
numerical_with_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype!='O']
```

```python
numerical_with_na
```

```python
for feature in numerical_with_na:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#Replacing the numerical missing values
for feature in numerical_with_na:
    #we will replace with median since there are outliers
    median_value=dataset[feature].median()
    #create new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
dataset[numerical_with_na].isnull().sum()    
```

# Handling rare categorical features

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(100)
```

```python
dataset.shape
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].median().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```
