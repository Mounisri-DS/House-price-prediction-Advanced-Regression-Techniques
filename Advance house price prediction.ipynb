```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import cross_val_score
```

```python
# Function to get cross validation scores
def get_cv_scores(model):
    scores=cross_val_score(model,train_X,train_Y,cv=5,scoring='r2')
    print('CV mean:',np.mean(scores))
    print('STD:',np.std(scores))
    print('\n')
```

```python
lr=LinearRegression().fit(train_X,train_Y)
get_cv_scores(lr)
```

Regularization can be defined as explicitly restricting a model to prevent overfitting.
Lets apply regularization to check whether the model is overfitted or not.

```python
alpha=1
ridge=Ridge(alpha=1).fit(train_X,train_Y)
get_cv_scores(ridge)
```

```python
# find optimal alpha with grid search
alpha=[0.001,0.01,0.1,1,10,100,1000]
param_grid=dict(alpha=alpha)
grid=GridSearchCV(estimator=ridge,param_grid=param_grid,scoring='r2',verbose=1,n_jobs=-1)
grid_result=grid.fit(train_X,train_Y)
print('Best Score:', grid_result.best_score_)
print('Best params:',grid_result.best_params_ )
```

```python
alpha=1
lasso=Lasso(alpha=1).fit(train_X,train_Y)
get_cv_scores(lasso)
```

```python
# find optimal alpha with grid search
alpha=[0.001,0.01,0.1,1,10,100,1000]
param_grid=dict(alpha=alpha)
grid=GridSearchCV(estimator=lasso,param_grid=param_grid,scoring='r2',verbose=1,n_jobs=-1)
grid_result=grid.fit(train_X,train_Y)
print('Best Score:', grid_result.best_score_)
print('Best params:',grid_result.best_params_ )
```

our score improved by optimizing alpha

Initially We got a r2 value of 0.8781 and standard deviation of 0.01674. The high value of r2 indicates our model is accurate and also it is clear that after applying ridge regularization our standard deviation little bit decreased which suggests we are less likely to be overfitting.
